```python
total_score_data = total_score_data[total_score_data['inning']==1]
total_score_data
```

```python
#merging this dataframe with match data
match.merge(total_score_data[['match_id','total_runs']],left_on = 'id',right_on='match_id')
```

```python
match_df = match.merge(total_score_data[['match_id','total_runs']],left_on = 'id',right_on='match_id')
```

```python
match_df['team1'].unique()
```

```python
#filtering out the teams and making common teams of the same name
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
# to remove dl matches
```

```python
match_df[['match_id','city','winner','total_runs']]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.merge(delivery,on='match_id')
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.shape
```

```python
delivery_df
```

```python
# Convert 'total_runs_y' to numeric if it is not already
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')

# Perform the cumulative sum operation within each group
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 120 - ((delivery_df['over']-1)*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
# Fill NaN values with "0" and convert player_dismissed to 0 or 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == "0" else 1)

# Convert to integer dtype
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)
```

```python
delivery_df['match_id'] = delivery_df['match_id'].astype(int)
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# finding current run rate
delivery_df['crr'] = delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
# finding required run rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# function to judge the winner
def result(row):
    return 1 if row['batting_team']== row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
delivery_df
```

```python
final_data = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
#shuffling the dataset to avoid bias 
final_df = final_data.sample(final_data.shape[0])
```

```python
final_df.sample()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0] # to remove infinite value at rrr
```

```python
final_df.shape
```

## Model Building

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# as our data still contains string we use one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# trying out random forest
from sklearn.ensemble import RandomForestClassifier
pipe_rf = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
```

```python
pipe_rf.fit(X_train,y_train)
```

```python
y_pred_rf = pipe_rf.predict(X_test)
accuracy_score(y_test,y_pred_rf)
```
